# Processing Calfornia Data using Nextdoor.com page

## Importing required Libraries

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from geopy.geocoders import Nominatim
```

```python
source = requests.get("https://nextdoor.com/find-neighborhood/ca/").text
```

```python
soup = BeautifulSoup(source,'lxml')
```

```python
soup
```

```python
html = soup.find_all("p")
```

```python
html
```

```python
len(html)
```

```python
# converting to bs4 result set to list of bs4 object

ul = list()

for item in html:
    ul.append(item.text)
    print(item.text)
```

```python
finalList = list()

for item in ul:
    if len(item) < 2:
        continue
    else:
        finalList.append(item)
```

```python
finalList
```

## Getting the Latitude and Longitude of the Neighborhood list

```python
dataFrame = pd.DataFrame(columns=['Neighborhood','Latitude','Longitude'])

geolocator = Nominatim()
i=0
for place in finalList:
    print(place)
    while True:
        try:
            geo = geolocator.geocode(place+" california")
            if geo != None:
                dataFrame.loc[i] = [place , geo.latitude, geo.longitude]
                i = i+1
            break
        except Exception as e:
            print(e)
```

## Saving the Data

```python
dataFrame.to_csv('final_California_csv')
```
